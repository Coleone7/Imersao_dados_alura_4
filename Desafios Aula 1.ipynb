```python
######################################################################################################################################### DESAFIO 1 e 3 
#1 - Realizar a média da metragem para cara um dos bairros (Paulo)
#3 - Explorar alguns gráficos na documentação e aplicar nas demais colunas do DF, tentar colocar alguma conclusão (Thiago)

metragem_bairro = dados.groupby('Bairro')['Metragem'].mean() 
metragem_bairro.head(20)
## média FEITA.
```

```python
################################################################################################################################################### Desafio 3

#4 - Pegar outros estatísticas dos dados, média, mediana, mim, max (Vivi)

media_b = dados["Banheiros"].mean()
print("Média de banheiros:", media_b)

mediana_b = dados["Banheiros"].median() 
print("Mediana de banheiros:", mediana_b)

max_b = dados["Banheiros"].max()
print("Máximo de banheiros em um imóvel:", max_b)

max_b = dados["Banheiros"].idxmax()  

min_b = dados["Banheiros"].min() 
print("Mínimo de banheiros num imóvel: ", min_b)

moda_b = dados["Banheiros"].mode()
print("Moda de banheiros:", moda_b)

```
